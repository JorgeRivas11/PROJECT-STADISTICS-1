```python
tabla['Promedio de satisfacción'] = df['Q1 Satisfaccion General'].mean().round(2)
print(tabla)
```

```python
tabla['Tasa de respuesta'] = round((108 / 481) * 100, 2)
print(tabla)
```

```python
prom = ((df["Q3 Recomendación"] >= 9) & (df["Q3 Recomendación"] <= 10)).sum()
det = (df["Q3 Recomendación"]<= 6).sum()
total_respuestas = df['Q3 Recomendación'].dropna().count()
nps = ((prom / total_respuestas) * 100) - ((det / total_respuestas) * 100)
print(total_respuestas)
print("NPS:", round(nps, 2))
print("Detractores:", det)
print("Promotores:", prom)
```

```python
prc_prom = round((prom / total_respuestas) * 100, 2)
prc_det = round((det / total_respuestas) * 100, 2)
print(f"Promotores: {prc_prom}%")
print(f"Detractores: {prc_det}%")
```

```python
tabla['Porcentaje detractores'] = prc_det
tabla['NPS'] = round(nps, 2)
tabla['Promotores'] = prom
tabla['Detractores'] = det
tabla['Dieron calificación de recomendación'] = total_respuestas
print(tabla.head())
```
